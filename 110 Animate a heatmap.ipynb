Paint a heatmap of the 50 random points.

```python
def get_heatmap(points):    image = np.zeros((256, 256))    for x, y in points:        image[y, x] += 1    return scipy.ndimage.gaussian_filter(image, (10, 10))pl.imshow(get_heatmap(points), origin='lower');
```
