```python
import pylab as pl

def plot_embedding(X):
    x_min, x_max = np.min(X, 0), np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)
    pl.scatter(X[:, 0], X[:, 1])
    pl.xticks([]), pl.yticks([])
    
from sklearn import manifold
visualization = manifold.LocallyLinearEmbedding(n_neighbors=30, n_components=2, method='standard')
```
