[Lasso](http://scikit-learn.org/dev/modules/linear_model.html#lasso)

```python
model = linear_model.Lasso(alpha=0.1)model.fit(vectors, targets)print model.coef_print model.intercept_
```

```python
# Adapted from# http://scikit-learn.org/dev/auto_examples/linear_model/plot_lasso_model_selection.htmlmodel = linear_model.LassoCV(cv=20)model.fit(diabetes.data, diabetes.target)
```

```python
def plot_model_selection(model, alphas, meanSquaredErrorPaths):    logAlphas = -np.log10(alphas)    pl.figure()    pl.plot(logAlphas, meanSquaredErrorPaths, ':')    pl.plot(logAlphas, meanSquaredErrorPaths.mean(axis=-1), 'k',        label='Average across folds', linewidth=2)    pl.axvline(-np.log10(model.alpha), linestyle='--', color='k',         label='alpha: CV estimate')    pl.legend()    pl.title('Mean square error on each fold: Coordinate descent')    pl.xlabel('-log(alpha)')    pl.ylabel('Mean square error')    pl.axis('tight')    pl.ylim(2300, 3800)    plot_model_selection(model, model.alphas, model.mse_path_)
```

[Lasso least angle regression](http://scikit-learn.org/dev/modules/linear_model.html#least-angle-regression)

```python
model = linear_model.LassoLars(alpha=0.1)model.fit(vectors, targets)print model.coef_print model.intercept_
```

```python
# Adapted from# http://scikit-learn.org/dev/auto_examples/linear_model/plot_lasso_model_selection.htmlmodel = linear_model.LassoLarsIC(criterion='bic')model.fit(diabetes.data, diabetes.target)pl.figure()pl.plot(-np.log10(model.alphas_), model.criterion_, '--', linewidth=3)pl.axvline(-np.log10(model.alpha_), linewidth=3)pl.xlabel('-log(alpha)')pl.ylabel('Information criterion')pl.title('Model selection by information criterion');
```
